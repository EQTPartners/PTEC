```python
import json
from sectors.experiments.nshot import results_to_df, plot_n_model_size, plot_method_model_size

paths = [
    "decapoda-research/llama-7b-hf/3.json",
    "bigscience/bloom-7b1/0.json",
    "bigscience/bloom-3b/0.json",
    "bigscience/bloom-1b7/0.json",
    "google/flan-t5-base/0.json",
    "google/flan-t5-large/0.json",
    "google/flan-t5-xl/3.json",
    "google/flan-t5-xxl/2.json",
    "gpt-3.5-turbo/0.json",
    "vicuna/7b/1.json",
]

allresults = []

for path in paths:
    with open(f"old_nshot_results/{path}", "r") as fp:
        allresults.append(json.load(fp))

results = results_to_df(allresults)
results.isna().sum().sum()
```
